```python
#Q2

import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")


def calculate_ave_1(points):
    average = []
    sum = np.cumsum(points)
    for i in range(sum.shape[0]):
        average.append(sum[i] / (i + 1))
    return average


def calcualate_ave_2(points, beta):
    average = []
    temp_average = 0
    for data in points:
        temp_average = temp_average * beta + data * (1 - beta)
        average.append(temp_average)
    return average


def calculate_ave_3(points, beta):
    average = []
    for item, data in enumerate(points):
        if item == 0:
            average.append(0)
            continue
        final_ave = data / (1 - beta ** item)
        average.append(final_ave)
    return average


def draw_graph(points, title):
  fig, axis = plt.subplots(1, 1, figsize=(20, 5))
  x = []
  y = []
  for item, data in enumerate(points):
    x.append(item)
    y.append(data)
  plt.plot(x, y)
  plt.title(title)
  plt.xlabel('Entry')
  plt.ylabel('Average')
  plt.show()

data = []
i = 0
for i in range(0,100):
  data.append(1)
  i = i + 1

for i in range(100,200):
  data.append(-1)
  i = i + 1


beta = 0.9
first_average_1 = calculate_ave_1(data)
second_average_1 = calcualate_ave_2(data, beta)
third_average_1 = calculate_ave_3(second_average_1, beta)

print("Beta = 0.9")
draw_graph(first_average_1, 1)
draw_graph(second_average_1, 2)
draw_graph(third_average_1, 3)


beta = 0.999
first_average_2 = calculate_ave_1(data)
second_average_2 = calcualate_ave_2(data, beta)
third_average_2 = calculate_ave_3(second_average_2, beta)

print("Beta = 0.999")
draw_graph(first_average_2, 1)
draw_graph(second_average_2, 2)
draw_graph(third_average_2, 3)
```
